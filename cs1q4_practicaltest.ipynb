```python
print("Welcome to Junluiz's Quiz Score Entry Program!")

user_name = input("Please enter your name: ")

num_scores = 0
while True:
    try:
        num_scores = int(input(f"\nHow many quiz scores will you enter? "))
        if num_scores > 0:
            break
        else:
            print("Please enter a positive number of scores.")
    except ValueError:
        print("Please enter a whole number.")

quiz_scores = []
print("\nEnter the scores of each quiz (from 0-100)")
for i in range(num_scores):
    while True:
        try:
            score = float(input(f"Quiz #{i + 1}: "))
            if 0 <= score <= 100:
                quiz_scores.append(score);
                break
            else:
                print("Score must be between 0 and 100.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def calculate_results(scores):
    total = sum(scores)
    if scores:
        average = total / len(scores)
    else:
        average = 0
    return total, average

total_score, average_score = calculate_results(quiz_scores)

print("\nHere is the summary! :)")
print(f"Scores for {user_name}:")
for i, score in enumerate(quiz_scores):
    print(f"  Quiz {i+1}: {score}")

print(f"Total score: {total_score:.2f}")
print(f"Average score: {average_score:.2f}")

if average_score >= 75:
    print("You PASSED! :D")
else:
    print("You FAILED.. :()")

print("Program finished. Goodbye friend!")
```
